```python
import os
train = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/train.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/train2.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/train3.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/train4.dat']
test =  ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/test.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/test2.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/test3.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/test4.dat']
outputs = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/tree.txt', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/tree2.txt', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/tree3.txt', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw1/tree4.txt']
# Put train files in .train/
!mkdir train 
for i in train:
    os.system('wget -P ./train ' + i + ' --no-check-certificate')
# Put test files in .test/
!mkdir test 
for i in test:
    os.system('wget -P ./test ' + i + ' --no-check-certificate')
# Put output files in ./output
!mkdir output
for i in outputs:
    os.system('wget -P ./output ' + i + ' --no-check-certificate')
```
