```python
import os
train = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/train.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/train2.dat']
test =  ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/test.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/test2.dat']
outputs = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/eval.txt', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw2/eval2.txt']
# Put train files in .train/
!mkdir train 
for i in train:
    os.system('wget -P ./train ' + i + ' --no-check-certificate')
# Put test files in .test/
!mkdir test 
for i in test:
    os.system('wget -P ./test ' + i + ' --no-check-certificate')
# Put output files in ./output
!mkdir output
for i in outputs:
    os.system('wget -P ./output ' + i + ' --no-check-certificate')
```
