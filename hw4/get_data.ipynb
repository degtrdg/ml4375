```python
import os
train = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/train4.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/train7.dat']
test =  ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/test4.dat', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/test7.dat']
outputs = ['https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t4l0n1a9i1000.out', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t4l1n2a9i1000.out', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t4l2n4a9i1000.out','https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t7l0n1a9i1000.out', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t7l1n2a9i1000.out', 'https://www.hlt.utdallas.edu/~vince/cs4375/assignments/hw4/t7l2n2a9i1000.out']
# Put train files in .train/
!mkdir train 
for i in train:
    os.system('wget -P ./train ' + i + ' --no-check-certificate')
# Put test files in .test/
!mkdir test 
for i in test:
    os.system('wget -P ./test ' + i + ' --no-check-certificate')
# Put output files in ./output
!mkdir output
for i in outputs:
    os.system('wget -P ./output ' + i + ' --no-check-certificate')
```
